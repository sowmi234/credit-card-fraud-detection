```python
import pandas as pd

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# Extract relevant datetime components as numerical features
df['year'] = df['trans_date_trans_time'].dt.year
df['month'] = df['trans_date_trans_time'].dt.month
df['day'] = df['trans_date_trans_time'].dt.day
df['hour'] = df['trans_date_trans_time'].dt.hour
df['minute'] = df['trans_date_trans_time'].dt.minute
df['second'] = df['trans_date_trans_time'].dt.second

```

```python
features = ['amt', 'lat', 'long', 'city_pop', 'hour', 'minute', 'second', 'year', 'month', 'day']
target='is_fraud'
```

```python
X_train, X_val, y_train, y_val = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_val)
```

```python
print("Accuracy:", accuracy_score(y_val, y_pred))
print("Classification Report:", classification_report(y_val, y_pred))
print("Confusion Matrix:", confusion_matrix(y_val, y_pred))
```
